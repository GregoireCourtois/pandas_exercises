### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep = '|')
```

### Step 4. Discover what is the mean age per occupation

```python
round(users.groupby(users.occupation).mean()[['age']])
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
mask = users.gender=='M'
male_users = users[mask]
male_users = male_users.groupby(male_users.occupation).count().gender

total_users = users.groupby(users.occupation).count().gender

ratio_users = round(male_users/total_users*100,2)
ratio_users.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(users.occupation).age.agg(['min','max']) #.agg function allows to apply multiple functions to the selected data
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
round(users.groupby(['occupation','gender']).mean()[['age']])
```
